```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt

%matplotlib inline
```

```python
df = pd.read_csv("loan_data.csv")
df.head()
```

## Data Visualization


```python
sns.countplot(df["Gender"], hue=df["Education"])
```

**There are Male who graduated and accepted thier loan status while Female is less educated less graduated**

```python
f, ax = plt.subplots(ncols=2, figsize=(13, 4))
plt.sca(ax[0])
sns.countplot(df["Gender"], hue=df["Married"], palette='rocket')

plt.sca(ax[1])
sns.countplot(df["Education"], hue=df["Married"], palette='rocket')
plt.show()
```

**still Male is highr in education and marrage than female even though Males who are not married and not educated still have high frequency than Femals**

- so most acceptable loan status is Male let\`s check this

```python
sns.countplot(df["Loan_Status"], hue=df["Gender"])
plt.show()
# as we can see most of our dataset of for Males
```

* Let\`t see the are where they comes from.

```python
sns.countplot(df["Property_Area"], hue=df["Gender"], palette="rocket")
# Semiurban!
```

```python
sns.countplot(df["Self_Employed"], hue=df["Loan_Status"])
plt.show()
```

**The one who accepted thier loan status is not self-employed this give us meaning they are rely almost on employed with other <span style="color:blue">(i.e not working with themself)</span>.**

```python
f, ax = plt.subplots(ncols=2, figsize=(13, 4))
plt.sca(ax[0])
sns.stripplot(x= df["Loan_Amount_Term"], y = df["LoanAmount"], dodge=True)

plt.sca(ax[1])
sns.stripplot(x= df["Credit_History"], y = df["LoanAmount"], dodge=True)
```

```python
f, ax = plt.subplots(ncols=2, figsize=(13, 4))
plt.sca(ax[0])
sns.swarmplot(data= df, x="Dependents", y ="ApplicantIncome", dodge=True)

plt.sca(ax[1])
sns.swarmplot(data= df, x="Dependents", y ="ApplicantIncome", dodge=True, hue='Education')
plt.legend(loc="upper center")
plt.ylabel("")

plt.show()
# we can see here application income is higher when they depends on 3 or more
# application incomes is higher when they graduated whether they dependends on indpendends
```

```python
f, ax = plt.subplots(ncols=2, figsize=(13, 4))
plt.sca(ax[0])
sns.swarmplot(data= df, x="Credit_History", y ="CoapplicantIncome", dodge=True)

plt.sca(ax[1])
sns.swarmplot(data= df, x="Credit_History", y ="CoapplicantIncome", dodge=True, hue='Property_Area')
plt.legend(loc="upper center")
plt.ylabel("")

plt.show()
# CoapplicantIncome higher and credit history is higher
```

```python
f, ax = plt.subplots(ncols=3, figsize=(15, 4))
plt.sca(ax[0])
sns.barplot(data=df, x="Gender", y = "Loan_Amount_Term")

plt.sca(ax[1])
sns.barplot(data=df, x="Gender", y = "CoapplicantIncome")

plt.sca(ax[2])
sns.barplot(data=df, x="Gender", y = "LoanAmount")

plt.show()

# Females is lower in CoapplicantIncome than Males!.
```

## Interpret boxplot

```python
sns.distplot(df["ApplicantIncome"], color='orange')
plt.show()
```

- Incomes seem to visualize outliers when its above 20,000, let\`s see the boxplot

```python
f, ax = plt.subplots(ncols=2, figsize=(13, 4))
plt.sca(ax[0])
sns.boxplot(df["ApplicantIncome"])

plt.sca(ax[1])
sns.violinplot(df["ApplicantIncome"], color="r")
# outliers when incomes above 20,000!

plt.show()
```
